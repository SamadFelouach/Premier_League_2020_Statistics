# Scraping 2011 Premier lEAGUE Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.skysports.com/premier-league-table/2019'
page=requests.get(url)
print(page.status_code)
```

```python
#print(page.text)
```

```python
soup= BeautifulSoup(page.text , 'html.parser')
```

```python
#print the first element 'a'
#print(soup.find('a'))
#print all the elements of 'a '
print(soup.find_all('a'))
```

```python
#find element by index
#soup.find_all('a')[1]
```

```python
league =soup.find('table' , class_ ='standing-table__table')
#print(league)
```

```python
league_table = league.find_all('tbody')
#print(league_table)
```

```python
league_2020 =[]

for teams_table in league_table:
        rows = teams_table.find_all('tr')
        for row in rows:
            team_names = row.find('td', class_='standing-table__cell--name').text.strip()
            team_points = row.find_all('td' , class_='standing-table__cell')[9].text.strip()
            league_dict={'name':team_names , 'points':team_points}
            league_2020.append(league_dict)
```
